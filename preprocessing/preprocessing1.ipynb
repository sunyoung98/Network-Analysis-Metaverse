```python
def data_text_cleaning(data):

    # ÏòÅÎ¨∏Ïûê Ïù¥Ïô∏ Î¨∏ÏûêÎäî Í≥µÎ∞±ÏúºÎ°ú Î≥ÄÌôò
    only_english = re.sub('[^a-zA-Z]', ' ', data)

    # ÏÜåÎ¨∏Ïûê Î≥ÄÌôò
    no_capitals = only_english.lower().split()

    # Î∂àÏö©Ïñ¥ Ï†úÍ±∞
    stops = set(stopwords.words('english'))
    no_stops = [word for word in no_capitals if not word in stops]

    # Ïñ¥Í∞Ñ Ï∂îÏ∂ú
    stemmer = nltk.stem.SnowballStemmer('english')
    stemmer_words = [stemmer.stem(word) for word in no_stops]

    # Í≥µÎ∞±ÏúºÎ°ú Íµ¨Î∂ÑÎêú Î¨∏ÏûêÏó¥Î°ú Í≤∞Ìï©ÌïòÏó¨ Í≤∞Í≥º Î∞òÌôò
    return ' '.join(stemmer_words)
```

```python
from multiprocessing import Pool

def use_multiprocess(func, iter, workers):
    pool = Pool(processes=workers)
    result = pool.map(func, iter)
    pool.close()
    return result
```

```python
import time
start_time = time.time()

zep_c2 = use_multiprocess(data_text_cleaning, zep_c,8 )
print('Ïã§Ìñâ ÏãúÍ∞Ñ :', (time.time() - start_time))
```
