```python
from nltk.stem import WordNetLemmatizer
```

```python
def data_text_cleaning(data):

    # ÏòÅÎ¨∏Ïûê Ïù¥Ïô∏ Î¨∏ÏûêÎäî Í≥µÎ∞±ÏúºÎ°ú Î≥ÄÌôò
    only_english = re.sub('[^a-zA-Z]', ' ', data)

    # ÏÜåÎ¨∏Ïûê Î≥ÄÌôò
    no_capitals = only_english.lower().split()
    
    #Ïä§ÌÖåÎ∞ç
    #stemmer = nltk.stem.SnowballStemmer('english')
    lemma = WordNetLemmatizer()
    lemma_words = [lemma.lemmatize(word) for word in no_capitals]
    
    #Î∂àÏö©Ïñ¥Ï†úÍ±∞
    stops = set(stopwords.words('english'))
    no_stops = [word for word in lemma_words if not word in stops]
    

    # Í≥µÎ∞±ÏúºÎ°ú Íµ¨Î∂ÑÎêú Î¨∏ÏûêÏó¥Î°ú Í≤∞Ìï©ÌïòÏó¨ Í≤∞Í≥º Î∞òÌôò
    return ' '.join(no_stops)
```

```python
from multiprocessing import Pool

def use_multiprocess(func, iter, workers):
    pool = Pool(processes=workers)
    result = pool.map(func, iter)
    pool.close()
    return result
```

```python
import time
start_time = time.time()

zep_c2 = use_multiprocess(data_text_cleaning, zep_c,8 )
print('Ïã§Ìñâ ÏãúÍ∞Ñ :', (time.time() - start_time))
```

```python
from nltk import word_tokenize,pos_tag
```

```python
tokens_docs=[word_tokenize(doc) for doc in zep_c2]
```

```python
len(tokens_docs)
```

```python
tokens_docs_tagged=[pos_tag(tokens_doc) for tokens_doc in tokens_docs]
```

```python
pos_of_interest=['NN','NNS','NNP','NNPS','VB','VBD','VBG','VBN','VBP','VBZ','JJ','JJR','JJS']
```

```python
filtered_tokens_docs_tagged=[[each for each in doc if (each[1] in pos_of_interest)] for doc in tokens_docs_tagged]
```

```python
len(filtered_tokens_docs_tagged)
```

```python
tokens_docs_selected=[docs for docs in filtered_tokens_docs_tagged if len(docs)>=5]
```

```python
len(tokens_docs_selected)
```

```python
from more_itertools import locate
ind_list = list(locate(filtered_tokens_docs_tagged, lambda x: len(x)>=5))

zep.iloc[ind_list,:].head()
```

```python
zep_selected=zep.iloc[ind_list,:]
zep_selected.shape
```

```python
len(tokens_docs_selected)
```

```python
words_docs_selected=[[word[0] for word in doc] for doc in tokens_docs_selected]
```
